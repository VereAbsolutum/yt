```python
import os

from PIL import Image
from config.app import PROJECT_NAME, IMAGE_FOLDER, IMAGE_FILENAME,OFFSET,MULTIPLIERS,APP,TMP,TMP_IMAGE_FOLDER

Image.MAX_IMAGE_PIXELS = 933120000

deslocamento = int(OFFSET)
tmp_img_folder = os.path.join(TMP, PROJECT_NAME, TMP_IMAGE_FOLDER)

begin = 2
#end = len(tmp_img_folder)
end = 10
os.listdir(tmp_img_folder)[begin:end]
```

```python
for cap, image in enumerate(os.listdir(tmp_img_folder)[begin:end], start=begin):
    filename = os.path.join(tmp_img_folder, image)
    img = Image.open(filename)
    largura, altura = img.size

    chatper = cap + 1
    cap_dir = os.path.join(APP, PROJECT_NAME, IMAGE_FOLDER, str(chatper))
    os.makedirs(cap_dir, exist_ok=True)
    
    for pos, inicio in enumerate(range(0, altura, deslocamento)):
        for cut_factor, fator in enumerate(MULTIPLIERS):
            filepath = os.path.join(cap_dir, f'{chatper:04d}-{pos:04d}-{cut_factor:04d}-{IMAGE_FILENAME}')
            fim = min(inicio + int(largura * fator), altura)
            recorte = img.crop((0, inicio, largura, fim))
            recorte.save(filepath)
            print('saved: ' + filepath)
```
